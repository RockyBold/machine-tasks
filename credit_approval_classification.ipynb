```python
col_map = {'CODE_GENDER':'Gender', 'FLAG_OWN_CAR':'Car', 'FLAG_OWN_REALTY':'Realty', 'CNT_CHILDREN':'Children',
           'AMT_INCOME_TOTAL':'Income', 'NAME_INCOME_TYPE':'Income_Type', 'NAME_EDUCATION_TYPE':'Education',
           'NAME_FAMILY_STATUS':'Family_Status', 'NAME_HOUSING_TYPE':'Housing', 'DAYS_BIRTH':'Age','DAYS_EMPLOYED':'Employment_Time',
          'FLAG_MOBIL':'Mobile', 'FLAG_WORK_PHONE':'Work_Phone', 'FLAG_PHONE':'Phone', 'FLAG_EMAIL':'Email', 'OCCUPATION_TYPE':'Occupation',
          'CNT_FAM_MEMBERS':'Fam_Members', 'dep_value':'Target'}
```

```python
n_data.rename(columns=col_map, inplace=True)
```

```python
n_data.head()
```

```python
n_data = n_data.drop(columns=['Education','Family_Status','Occupation','Email','Phone','Mobile',
                              'Work_Phone','Gender','Employment_Time', 'ID', 'Children', 'Age'])
```

```python
n_data.nunique()
```

```python
n_data.loc[n_data['Car']=='Y', 'Car'] = 1
n_data.loc[n_data['Car']=='N', 'Car'] = 0
n_data['Car'] = n_data['Car'].astype('int')

n_data.loc[n_data['Realty']=='Y', 'Realty'] = 1
n_data.loc[n_data['Realty']=='N', 'Realty'] = 0
n_data['Realty'] = n_data['Realty'].astype('int')

n_data.loc[n_data['Target']=='Yes', 'Target'] = 1
n_data.loc[n_data['Target']=='No', 'Target'] = 0
n_data['Target'] = n_data['Target'].astype('int')
```

```python
n_data['Age'] = (n_data['Age']*-1)/365
n_data['Age'] = scaling(n_data, 'Age')
```

```python
n_data.dtypes
```

```python
n_data = pd.get_dummies(n_data)
```

```python
n_data['Income'] = scaling(n_data,'Income')
```

```python
x = n_data.drop(columns=['Target'])
y = n_data['Target']
```

```python
def scaling(df, col):
    df[col] = (df[col]-df[col].mean())/df[col].std()
    return df[col]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,shuffle=True)
```

```python
x_train
```

```python
dec_tree = DecisionTreeClassifier()
dec_tree.fit(x_train,y_train)
```

```python
y_predictions_dt = dec_tree.predict(x_test)
```

```python
metrices = {'Accuracy': [accuracy_score(y_test, y_predictions_dt)]
            ,'Precision':[precision_score(y_test, y_predictions_dt)]
            ,'Recall':[recall_score(y_test, y_predictions_dt)]
            ,'F1_score':[f1_score(y_test, y_predictions_dt)]}
dt_scores = pd.DataFrame(data=metrices)
dt_scores
```
