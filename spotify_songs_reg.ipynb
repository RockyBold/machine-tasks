```python
df = df[['danceability','energy','key','loudness','mode','speechiness','acousticness','instrumentalness','liveness','valence','tempo','playlist_genre']]
```

```python
le = LabelEncoder()
df['playlist_genre'] = le.fit_transform(df['playlist_genre'])
```

```python
def normalise(df, cols=[]):
    for col in cols:
        df[col] = (df[col]-df[col].min())/(df[col].max()-df[col].min())
    return df
```

```python
x = df.drop(columns=['valence'])
x = sm.add_constant(x)
y = df['valence']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True)
```
